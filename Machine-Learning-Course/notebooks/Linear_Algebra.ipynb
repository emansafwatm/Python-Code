**Multiplying Matrices**

We can only multiply two matrices if the number of rows in matrix A is the same as the number of columns in matrix B.

In another way we can say wo matrices are compatible for multiplication if the number of columns of 1 matrix is equal to the number of rows of the other matrix.

For example, if matrix 1 has dimensions a * N and matrix 2 has dimensions N * b, then the resulting matrix has dimensions of a * b.

image.png

image.png

image.png

```python
A = np.array([[1, 2,3],[4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
C= np.dot(A,B)
C
```
